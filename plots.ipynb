```python
# Accuracy vs communication

import matplotlib.pyplot as plt
import numpy as np

comm = np.array([30,
                 62,
                 88,
                 19,
                 35,
                 45, 
                 12])

acc = np.array([0.92 , 0.94  , 0.94,
    0.92 , 0.95  , 0.95,
    0.92 , 0.95  , 0.95,
    0.92 , 0.94  , 0.96,
    0.92 , 0.95  , 0.97,
    0.92 , 0.95  , 0.98,
    0.92 , 0.95  , 0.92])

mem = np.array([ 859 , 
                1654 ,
                2449 ,
                859  ,
                1654 ,
                2449 ,
                859  ,
                1654 ,
                2449 ])


idx = np.argsort(comm)

acc = np.array(acc).reshape(7,-1)

fig, ax1 = plt.subplots()

# plt.figure(figsize=(10,10))
ax1.plot(comm[idx],acc[idx,::-1],'o-')
ax1.set_ylim([0.8,1])
ax1.set_ylabel("Accuracy")
ax1.set_xlabel("Communication")
ax1.legend(['Overall', 'Cloud', 'Local'], loc=4)

ax2 = ax1.twinx()
ax2.set_ylim([800,4000])
ax2.plot(comm[idx], mem[idx],'*--c')
ax2.set_ylabel('Memory', color='c')

fig.savefig("_figures/commvsacc.pdf")
```

```python
# Accuracy vs communication

import matplotlib.pyplot as plt
import numpy as np

size = [0.1,
        0.5,
        0.8,
        1.0]

exit = [0.00,
        1.75,
        60.82,
        100.00]
        
acc = [ 0.92   ,  0.96  ,  0.96,
        0.92   ,  0.96  ,  0.96,
        0.92   ,  0.96  ,  0.97,
        0.92   ,  0.96  ,  0.92 ]
acc = np.array(acc).reshape(4,-1)

fig, ax1 = plt.subplots()

# plt.figure(figsize=(10,10))
ax1.plot(size,acc[:,::-1],'o-')
ax1.set_ylim([0.8,1])
ax1.set_ylabel("Accuracy")
ax1.set_xlabel("Threshold")
ax1.legend(['Overall', 'Cloud', 'Local'], loc="best")

ax2 = ax1.twinx()
ax2.set_ylim([0,200])
ax2.plot(size,exit,'*--c')
ax2.set_ylabel('% Locally Exited', color='c')

fig.savefig("_figures/thresholdvsacc.pdf")
```
